Triangular pattern generator

Using *

```python
rows=6
for i in range (1,rows+1):
  for j in range (i):
   print("*",end=" ")
  print()
```

Using numbers

```python
rows = 7
for i in range(1,rows +1):
  for j in range (i):
   print(++i,end=" ")
  print()
```

Character type identifier

```python
ch = input("Enter a character: ")

if ch.isalpha():
    print("Alphabet")
elif ch.isdigit():
    print("Digit")
else:
    print("Special Character")
```

Fibonacci sequence using while loop

```python
n =int(input("Enter a num. here:"))
a=0
b=1
count=0
while count<n:
 print(a,end=" ")
 c=a+b
 a=b
 b=c
 count+=1
```

Factorial generator using loop

```python
num=int(input("enter no. here:"))
if num <0:
 print("error")
elif num==0:
 print("1")
else:
 factorial=1
 for i in range (1, num + 1):
  factorial=factorial*i
  print("factorial:", factorial)
```
