Area's of Geometric figure

```python
print("To find the area of circle,press 1")
print("To find the area of triangle,press 2")
print("To find the area of rectangle,press 3")
a=int(input())

if a==1 :
  print("Enter circle's radius:")
  r=int(input())
  print(3.14*r*r,)

elif a==2 :
  print("Enter triangle's base:")
  b=int(input())
  print("Enter Triangle's height:")
  h=int(input())
  print((1/2)*h*b)

elif a==3 :
  print(" enter side 1 of rectangle:")
  s1=int(input())
  print(" enter side 2 of rectangle:")
  s2=int(input())
  print(s1*s2)

else:
  print("error")
```

Conversion Utilities

```python


print("To calculate degree Celsius to Fahrenheit,press 1")
print("To calculate Fahrenheit to degree Celsius,press 2")
print("To calculate degree Celsius to kelvin,press 3")
print("To calculate kelvin to degree Celsius,press 4")
a=int(input())
if a==1 :
  print("Enter Temperature  in celcius to get it in Fahrenheit:")
  c1=int(input())
  print((9*c1/5)+32)
elif a==2 :
  print("Enter Temperature in Fahrenheit to get it in celcius:")
  f1=int(input())
  print(((f1-32)*5)/9)
elif a==3 :
  print("Enter Temperature in celcius to get it in kelvin:")
  c2=int(input())
  print(c2-273)
elif a==4 :
  print("Enter Temperature in Kelvin to get it in celcius :")
  k1=int(input())
  print(k1+273)
else:
  print("error")
```

Gross salary of employee

```python
print("Enter your in hand salary:")
sal=int(input())

print("Enter your percentage of tax:")
tax=int(input())
print((sal*100)/(100-tax))

print("Enter your overall bonus:")
b=int(input())
print(((sal*100)/(100-tax))+b)
```

Calculator for simple interest

```python
print("Enter the amount of loan taken:")
l=int(input())
print("Enter the number of years loan's taken for: ")
n=int(input())
print("Enter the rate at which loan is taken at :")
r=int(input())
print("The total amount to be paid is")
print((l*n*r/100)+l)
print("and intrest will be ")
print(l*n*r/100)
```

Arithmetic operations

```python
print("To Add,press 1")
print("To Substract,press 2")
print("To Multiply,press 3")
print("To Divide,press 4")
a=int(input())

if a==1 :
  print("Enter number 1 to add:")
  b=int(input())
  print("Enter number 2 to add:")
  c=int(input())
  print(b+c)

elif a==2 :
  print("Enter number to Substract from:")
  d=int(input())
  print("Enter number to substract:")
  e=int(input())
  print(d-e)
elif a==3 :
  print("Enter number 1 to multiply:")
  f=int(input())
  print("Enter number 2 to multiply:")
  g=int(input())
  print(f*g)

elif a==4 :
  print("Enter number to be divided:")
  h=int(input())
  print("Enter number to divide:")
  i=int(input())
  print(h/i)

else:
  print("error")
```
